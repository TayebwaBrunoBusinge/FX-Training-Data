```python
import os
import pandas as pd

# List of directories containing csv files
dir_list = [
    # "/media/bustabruno/f71fc4fd-abcd-4a4b-9fb1-6a8899041c1c/FX-Training-Data/Sorted/EURUSD",
    # "/media/bustabruno/f71fc4fd-abcd-4a4b-9fb1-6a8899041c1c/FX-Training-Data/Sorted/GBPUSD",
    "/media/bustabruno/f71fc4fd-abcd-4a4b-9fb1-6a8899041c1c/FX-Training-Data/Sorted/GBPJPY",
]

# Iterate over the directories
for dir_name in dir_list:
    # List all the csv files in the directory
    csv_files = [f for f in os.listdir(dir_name) if f.endswith("_modified.csv")]

    # Iterate over the csv files
    for csv_file in csv_files:
        # Load the csv file into a pandas dataframe
        df = pd.read_csv(os.path.join(dir_name, csv_file))

        # Create a new column called "nxt_move_1H" and set it to 0
        df["nxt_move_1H"] = 0

        # Iterate over the rows of the dataframe
        for i, row in df.iterrows():
            # Check if iteration is at the last row.
            if i+1 == len(df):
              break
            # Check if the current candle closed bullish
            if row["st_bullish"] == 1:
                # Calculate the price movement in pips in the next hour
                # by subtracting the current high price from the next high price
                nxt_move_1H = (df.loc[i+1, "p_high"] - row["p_high"]) * 100
            # Check if the current candle closed bearish
            elif row["st_bearish"] == 1:
                # Calculate the price movement in pips in the next hour
                # by subtracting the current low price from the next low price
                nxt_move_1H = (df.loc[i+1, "p_low"] - row["p_low"]) * 100
            # If the current candle closed neutral, use the state of the former candle
            else:
                # Check if the previous candle closed bullish
                if df.loc[i-1, "st_bullish"] == 1:
                    # Calculate the price movement in pips in the next hour
                    # by subtracting the current high price from the next high price
                    nxt_move_1H = (df.loc[i+1, "p_high"] - row["p_high"]) * 100
                # Check if the previous candle closed bearish
                elif df.loc[i-1, "st_bearish"] == 1:
                    # Calculate the price movement in pips in the next hour
                    # by subtracting the current low price from the next low price
                    nxt_move_1H = (df.loc[i+1, "p_low"] - row["p_low"]) * 100
                # If the previous candle closed neutral, use the close prices
                else:
                    # Calculate the price movement in pips in the next hour
                    # by subtracting the current close price from the next close price
                    nxt_move_1H = (df.loc[i+1, "p_close"] - row["p_close"]) * 100
            
            # Set the value in the new column
            df.loc[i, "nxt_move_1H"] = round(nxt_move_1H,2)

        # Save the dataframe to a new csv file
        df.to_csv(os.path.join(os.path.join(dir_name, csv_file)), index=False)
```
